1.iris数据集聚类

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
```

```python
X=iris.data
y=iris.target
```

k-means 聚类，循环k=1-15，选择sse最小的k

```python
clusters=range(1,15)
sse=[]

for k in clusters:
    model=KMeans(n_clusters=k)
    model.fit(X)
    sse.append(model.inertia_) #返回评价指标 越小 聚类效果越好
```

```python
sse
```

```python
plt.plot(clusters, sse)
plt.xlabel('Number of clusters')
plt.ylabel('sse')
```

```python
###选择k=2聚类获得类标签
model2=KMeans(n_clusters=2)
y_pred=model2.fit_predict(X) 
y_pred
```

```python
###因为data2中有4个列属性，我们没法在四维空间中观察，因此，下面将使用降维方法（主成分分析PCA），从4维降到2维度，这2个维度是由4个维度新生成的，并不是从4个维度去掉2个维度。
from sklearn.decomposition import PCA 
pca = PCA(2) # 降维为2个维度
X_pca=pca.fit_transform(X) 
plt.scatter(x=X_pca[:,0], y=X_pca[:,1], c=y_pred) 
```

```python
###选择k=3聚类获得类标签
model3=KMeans(n_clusters=3)
y_pred=model3.fit_predict(X) 
plt.scatter(x=X_pca[:,0], y=X_pca[:,1], c=y_pred) 
```

虽然k=3会使得sse更小，但是从图上可以看出选择k=2是比较符合自然分布的。

2.图像压缩

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import os
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('2.jpg') #BGR通道
plt.imshow(img) ###图片显示不正常，因为颜色空间不对，需要做一下转换
```

```python
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #转换为RGB
plt.imshow(img1)
```

```python
img1.shape ###前两个维度是尺寸，第三个数表示3通道BGR
```

```python
dat=img1.reshape((4066*7713,3)) ###将图像的shape转换为350*525行，3列，即常见的样本形式，一行是一个样本，一列是一个属性。
dat.shape
```

```python
###使用Kmeans聚类
kmeans = KMeans(n_clusters=60).fit(dat) ###16个类
```

```python
kmeans.cluster_centers_ ###输出类中心
```

```python
len(kmeans.labels_) ###很长
```

```python
kmeans.labels_[0:10] ###显示前10个看看，猜测是每个样本的类标号 前十个像素属于哪个簇
```

```python
np.unique(kmeans.labels_) ####果然是16个类。验证了上面的猜测。
```

下面，写一个循环，将每个样本以类中心代替。

```python
label = []
for i in kmeans.labels_: ###kmeans.labels_是所有样本的类标号
    label.append(list(kmeans.cluster_centers_[i,:])) ###cluster_centers_是所有类的类中心
```

```python
label[0:10]
```

```python
new_img = np.array(label)
new_img = new_img.reshape((4066, 7713, 3)) ###再从样本的形式转换为图像的形式
plt.imshow(new_img)
```

上面提示了出错的原因，说明要么把new_img归一化到[0,1]保留小数，要么将其在[0,255]范围取整

```python
new_img1=new_img.astype(int) #整形 去掉小数点
plt.imshow(new_img1)
#new_img1.shape
```
