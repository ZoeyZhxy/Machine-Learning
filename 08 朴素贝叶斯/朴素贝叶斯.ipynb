```python
import pandas as pd
```

```python
#建立文本数据
documents=[['菜品','很','一般','不','建议','在这','消费'],
          ['老板','很','闹心','坑人','建议','去','别家'],
          ['让人','惊艳','东西','口味','让人','感觉','不错'],
          ['环境','不错','自然牛柳','很','好吃'],
          ['味道','真的','一般般','环境','也','比较','拥挤'],
          ['一家','性价比','很','高','餐厅','推荐']] 
```

```python
classVec=[1,1,0,0,1,0]
```

```python
words_all=set() #建立字典
```

```python
#创建词集合
for document in documents:
    words_all=words_all|set(document) #并运算，并且没有重复
```

```python
words_all
```

```python
#将词集合转化为列表
words_all=list(words_all)
```

```python
words_all
```

```python
dic={}
```

```python
document=documents[0]
```

```python
for word in words_all
```

```python
def creat_wordVec(document,words_all):
    dic={}
    for word in words_all:
        if word in document:
            dic[word]=1
        else:
            dic[word]=0
    return dic
```

```python
trainMatrix=[]
```

```python
for document in documents:
    trainMatrix.append(creat_wordVec(document,words_all))
```

```python
trainMatrix
```

```python
df=pd.DataFrame(trainMatrix)
```

```python
df
```

```python
#将标签转换为序列
se=pd.Series(classVec)
#取消极词的数据
df_neg=df[se==1]
```

```python
se
```

```python
df_neg
```

```python
#词出现的次数
p_negWordNum=df_neg.sum(axis=0)
p_negWordNum
```

```python
#消极词的总数
p_negAllNmum=p_negWordNum.sum()
p_negAllNmum
```

```python
#消极词的条件概率
p_negVect=p_negWordNum/p_negAllNmum
p_negVect
```

```python
p_negDoc=len(df_neg)/len(df)
p_negDoc
```

```python
df_pos=df[se==0]
p_posWordNum=df_pos.sum(axis=0)
p_posAllNmum=p_posWordNum.sum()
p_posVect=p_posWordNum/p_posAllNmum
p_posDoc=len(df_pos)/len(df)
```

```python
print(p_posVect)
print(p_posDoc)
```

```python
testDoc=['环境','很','不错']
```

```python
p_negCurVect=p_negVect[testDoc]
p_negCurVect
```

```python
p_negCla=p_negDoc
```

```python
#计算总概率
p_negFeatCla=p_negCurVect.prod()
p_negFeatCla
```

```python
p_posCurVect=p_posVect[testDoc]
p_posCurVect
```

```python
p_posFeatCla=p_posCurVect.prod()
p_posFeatCla
```

```python
p_posCla=1-p_negCla
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
wines=datasets.load_wine()
X=wines['data']
y=wines['target']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
gnb.score(X_test,y_test)
```
