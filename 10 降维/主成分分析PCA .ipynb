```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris=load_iris()
```

```python
X=iris.data
```

```python
y=iris.target
```

```python
X.shape
```

```python
n_dim=2 #降到2维
```

```python
pca=PCA(n_components=n_dim)
```

```python
X_pca=pca.fit_transform(X) #降维（一步）
```

```python
X_pca
```

```python
colors=['red','blue','green']
```

```python
for color,i, target_name in zip(colors,[0,1,2],iris.target_names): #分三次画
    plt.scatter(X_pca[y==i,0],X_pca[y==i,1],color=color,label=target_name) #X_pca[y==i,0] y==0为True取行 类别为0的 第一维和第二维
```

```python
from sklearn.cluster import KMeans
```

```python
a=np.array([[1,2,2,2,3],[2,1,2,3,2]])
b=np.array([[5,5,6,6,7,7],[6,7,6,7,6,7]])
```

```python
a
```

```python
b
```

```python
X=np.hstack((a,b)).T #把a和b堆起来
X
```

```python
y_pred=KMeans(n_clusters=2) #两簇
```

```python
y_pred.fit_predict(X) #表示簇标号
```
