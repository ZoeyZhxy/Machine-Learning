```python
from sklearn.neural_network import MLPClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
iris=datasets.load_iris()
```

```python
X=iris['data']
y=iris['target']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
clf=MLPClassifier(solver='adam',alpha=1e-5,hidden_layer_sizes=(3,3),random_state=1,max_iter=10000,activation='logistic') #solver梯度下降的方法
```

```python
clf.fit(X_train,y_train)
```

```python
clf.predict(X_test) 
```

```python
y_test
```

```python
clf.score(X_test,y_test)
```

使用模拟的数据集

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
```

```python
X=np.linspace(-3.14,3.14,400)
```

```python
X
```

```python
X1=X.reshape(-1,1) #把X变成二维 a.reshape(-1, x)则是将矩阵a变成列数为x，行数不规定的矩阵，具体行数按照总元素个数除列数，均分得到
```

```python
X1
```

```python
y=np.sin(X)+0.3*np.random.rand(len(X)) 
```

```python
reg=MLPRegressor(alpha=1e-5,hidden_layer_sizes=(3,2),random_state=1,max_iter=10000,activation='logistic') #(3,2) 3层每层2个神经元 隐藏层的激活函数，不是输出层
```

```python
reg.fit(X1,y)
```

```python
y_pred=reg.predict(X1) 
```

```python
y_pred
```

```python
y
```

```python
plt.scatter(X,y,c='red')
plt.plot(X,y_pred)
```

```python
reg.score(X1,y)
```

糖尿病数据集

```python
diabetes=datasets.load_diabetes()
```

```python
X=diabetes['data']
y=diabetes['target']
```

```python
X.shape #442个特征，有10个特征
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
reg=MLPRegressor(alpha=1e-6,hidden_layer_sizes=(10,4),random_state=1,max_iter=10000)
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test) 
```

```python
y_pred
```

```python
y_test
```

```python
reg.score(X_test,y_test) #做了归一化
```
