```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model, datasets
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
```

```python
diabetes=datasets.load_diabetes()
```

```python
y=diabetes['target']
```

```python
x=diabetes['data'][:,2] #取bmi指标
#x=diabetes['data']
```

```python
#x.shape 会报错（一维数据无法被检测到，需要二维数据）
```

```python
x=x.reshape(442,1) 
```

```python
x_train=x[:-20]
```

```python
x_test=x[-20:]
print(x_test)
```

```python
y_train=y[:-20]
y_test=y[-20:]
print(y_test)
```

```python
reg=linear_model.LinearRegression() #x新建线性回归模型
reg.fit(x_train,y_train) #训练模型
y_pred=reg.predict(x_test) #y hat
```

```python
print(reg.coef_) #w的值 斜率
```

```python
print(reg.intercept_) #b的值 截距
```
