```python
def ty(k,b,p):
    x=range(0,12)
    y=[k*i+b for i in x]#列表
    x1=(k*(p[0]-b)+p[0])/(k*k+1)
    y1=k*x1+b
    return {'line':[x,y],'tyd':[x1,y1]}
```

```python
k=0.6
b=0
p=[5,5]
```

```python
data=ty(k,b,p)
```

```python
x=data['line'][0]
y=data['line'][1]
x1=data['tyd'][0]
y1=data['tyd'][1]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
```

```python
iris=datasets.load_iris()
```

```python
X=iris['data']
y=iris['target']
```

```python
target_names=iris['target_names']
```

```python
plt.scatter(X[y==0,0],X[y==0,1],marker='v',s=30,c='r') #y==0，第一类的X里的第一个特征和第二个特征
plt.scatter(X[y==1,0],X[y==1,1],marker='o',s=30,c='k')
#plt.scatter(X[y==2,0],X[y==2,1],marker='p',s=30,c='b')
plt.legend(iris['target_names'][0:2])
plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])
```

```python
X=X[(y==0)|(y==1),0:2]
y=y[(y==0)|(y==1)]
```

```python
lda=LinearDiscriminantAnalysis(n_components=1) #压缩成1维
```

```python
ld=lda.fit(X,y)
```

```python
X_t=ld.transform(X)
```

```python
y_t=np.zeros(X_t.shape) #生成100个0 括号里是0的个数
y_t
```

```python
plt.scatter(X_t[y==0],y_t[y==0],marker='v',s=30,c='r')
plt.scatter(X_t[y==1],y_t[y==1],marker='o',s=30,c='k')
plt.lengend(iris['target_names'][0:2])
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
ld=lda.fit(X_train,y_train)
```

```python
y_pred=ld.predict(X_test) 
```

```python
list(zip(y_pred,y_test,y_pred==y_test))
```

```python
ld.score(X_test,y_test)
```
